```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.feature_extraction.text import TfidfVectorizer

# Load dataset
df = pd.read_csv(r"C:\Users\manee\Downloads\spam.csv", encoding='latin-1')

# Keep only required columns
df = df[['v1', 'v2']]
df.columns = ['label', 'message']

# Convert labels to binary
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

print(df.head())

tfidf = TfidfVectorizer(stop_words='english')

X = tfidf.fit_transform(df['message'])
y = df['label']

# Total number of features
total_features = X.shape[1]
print("Total TF-IDF Features:", total_features)

lasso = Lasso(alpha=0.1)
lasso.fit(X.toarray(), y)
coefficients = lasso.coef_

non_zero = np.sum(coefficients != 0)
zero = np.sum(coefficients == 0)

print("Non-zero coefficients (selected features):", non_zero)
print("Zero coefficients (eliminated features):", zero)
alphas = [0.01, 0.1, 1]
results = {}
for a in alphas:
    model = Lasso(alpha=a)
    model.fit(X.toarray(), y)    
    coef = model.coef_
    selected = np.sum(coef != 0)    
    results[a] = selected

for k, v in results.items():
    print(f"Alpha = {k} → Selected Features = {v}")
reduction = ((total_features - non_zero) / total_features) * 100

print("Percentage Reduction:", reduction, "%")
```
